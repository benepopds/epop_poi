```python
#rlist=rstt_temp.name_x.unique().tolist()
#print(rlist[1])
for i in range(0,len(rlist)):
    aa= rstt_temp[rstt_temp['name_x']==rlist[i]]['COMPANY_ID'].tolist()
    #print(aa)
    conn = pymysql.connect(host='192.168.0.50', user='eums', password='eums00!q', db='eums', charset='utf8')
    curs = conn.cursor(pymysql.cursors.DictCursor)
    curs.execute("select sum(PAYMENT) as Q1 from MEUMS_COMP_REVISIT_HIS WHERE COMPANY_ID IN "+str(tuple(aa))+" AND deal_dt between '2016-01-01 00:00:00' and '2016-03-31 23:59:59';")
    p1 = pd.DataFrame( curs.fetchall() )
    #p = curs.fetchall()
    curs.execute("select sum(PAYMENT) as Q2 from MEUMS_COMP_REVISIT_HIS WHERE COMPANY_ID IN "+str(tuple(aa))+" AND deal_dt between '2016-04-01 00:00:00' and '2016-06-31 23:59:59';")
    p2 = pd.DataFrame( curs.fetchall() )
    curs.execute("select sum(PAYMENT) as Q3 from MEUMS_COMP_REVISIT_HIS WHERE COMPANY_ID IN "+str(tuple(aa))+" AND deal_dt between '2016-07-01 00:00:00' and '2016-09-31 23:59:59';")
    p3 = pd.DataFrame( curs.fetchall() )
    curs.execute("select sum(PAYMENT) as Q4 from MEUMS_COMP_REVISIT_HIS WHERE COMPANY_ID IN "+str(tuple(aa))+" AND deal_dt between '2016-10-01 00:00:00' and '2016-12-31 23:59:59';")
    p4 = pd.DataFrame( curs.fetchall() )
    curs.execute("select sum(PAYMENT) as Q5 from MEUMS_COMP_REVISIT_HIS WHERE COMPANY_ID IN "+str(tuple(aa))+" AND deal_dt between '2017-01-01 00:00:00' and '2017-03-31 23:59:59';")
    p5 = pd.DataFrame( curs.fetchall() )
    bb = pd.concat([p1,p2,p3,p4,p5],axis=1)
    conn.close()
    bb['name']=rlist[i]
    rstt_sales=rstt_sales.append(bb)
rstt_sales
```
