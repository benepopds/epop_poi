```python
p = pd.DataFrame([])
p = p.append(p1)
p = p.append(p2)
p = p.append(p3)
p = p.append(p4)
print(p.shape)
```

```python
p['EPOPCON_ID'] = p['EPOPCON_ID'].str[5:50]
p.head()
```

```python
#전체 column 기준 groupby
p_group = p.groupby(['EPOPCON_ID','COMPANY_ID','CARD_NAME']).agg({'CNT_PAY':'sum','SUM_PAY':'sum'}).reset_index()
print(p_group.shape)
p_group.head()
```

```python
p_ex = p[['EPOPCON_ID','COMPANY_ID', 'CARD_NAME','LATITUDE1', 'LONGITUDE1', 'CO_NAME', 'CATE', 'CATE1', 'ADDR']]
print(p_ex.shape)
p_ex.head()
```

```python
hana = pd.merge(p_group, p_ex, how='inner', on=['EPOPCON_ID','COMPANY_ID','CARD_NAME'])
print(hana.shape)
hana.head()
```
